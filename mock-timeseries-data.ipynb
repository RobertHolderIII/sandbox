```python
import pandas as pd

series = pd.Series(range(0,1000))
```

```python
from math import sin, cos
xs = range(1000)
trend = [0.01*x for x in xs]
seasonality = [sin(0.1*x) + 0.3*cos(0.2*x) for x in xs]
```

```python
import matplotlib.pyplot as plt
plt.plot(xs, trend)
```

```python
plt.plot(xs, seasonality)
```

```python
plt.plot(xs, [sum(x) for x in zip(trend,seasonality)])
```

```python
import random
noise = [0.75*random.random() for _ in xs]
plt.plot(xs, noise)
```

```python
plt.plot(xs, [sum(x) for x in zip(trend,seasonality,noise)])
```

```python
import random
noise2 = [random.random()*random.random() for _ in xs]
plt.plot(xs, noise2)
```

```python
plt.plot(xs, [sum(x) for x in zip(trend,seasonality,noise2)])
```

```python
import random
noise3 = [random.random()*random.random()*random.random() for _ in xs]
plt.plot(xs, noise3)
```

```python
plt.plot(xs, [sum(x) for x in zip(trend,seasonality,[2*i for i in noise3])])
```

```python
import random
noise4 = [i-0.5 for i in noise3]
plt.plot(xs, noise4)
```

```python
plt.plot(xs, [sum(x) for x in zip(trend,seasonality,noise4)])
```
